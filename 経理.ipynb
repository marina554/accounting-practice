```python
# 区分（カテゴリ）と支払方法をもとに勘定科目を自動判定

def get_accounts(category, payment_method):
    # 区分（借方）を会計っぽくマッピング
    debit_map = {
        "売上": "売上高",
        "仕入": "仕入高",
        "事務用品費": "消耗品費",
        "旅費交通費": "旅費交通費",
        "通信費": "通信費",
        "地代家賃": "地代家賃",
        "水道光熱費": "水道光熱費",
        "交際費": "交際費",
        "広告宣伝費": "広告宣伝費",
        "雑費": "雑費"
    }

    # 借方勘定科目
    debit = debit_map.get(category, "不明")

    # 支払方法による貸方勘定科目の決定
    credit_map = {
        "現金": "現金",
        "銀行振込": "普通預金",
        "クレジットカード": "未払金",
        "掛け": "買掛金",
        "売掛": "売掛金"
    }

    credit = credit_map.get(payment_method, "不明")

    # 売上のように、取引の方向が逆になるケースを処理
    # （売上は貸方が売上高になる）
    if category == "売上":
        debit, credit = credit, debit  # 借貸を入れ替える

    return debit, credit
# 関数を実行して結果を表示
debit, credit = get_accounts("事務用品費", "現金")
print("借方:", debit)
print("貸方:", credit)
```
