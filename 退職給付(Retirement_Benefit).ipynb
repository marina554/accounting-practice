```python
import pandas as pd
import matplotlib.pyplot as plt

# サンプル従業員データ
employees = pd.DataFrame({
    'Employee': ['A', 'B', 'C'],
    'Salary': [5000, 6000, 5500],   # 年俸
    'YearsOfService': [10, 15, 20],
})

# 前提条件
discount_rate = 0.03   # 割引率
expected_return = 0.05 # 予定利率
retirement_benefit_per_year = 100 # 勤続年1年あたりの退職給付

# 退職給付債務(PBO)計算
employees['PBO'] = employees['YearsOfService'] * retirement_benefit_per_year / ((1 + discount_rate) ** employees['YearsOfService'])

# 年度ごとの費用計算（簡易）
employees['Expense'] = employees['Salary'] * expected_return

# 総額集計
total_pbo = employees['PBO'].sum()
total_expense = employees['Expense'].sum()

print("Total PBO:", total_pbo)
print("Total Expense:", total_expense)

# グラフ化
plt.bar(employees['Employee'], employees['PBO'], color='skyblue', label='PBO')
plt.bar(employees['Employee'], employees['Expense'], color='orange', alpha=0.7, label='Expense')
plt.ylabel('Amount')
plt.title('Retirement Benefit Portfolio')
plt.legend()
plt.show()
```
