```python
# Function to calculate profit or loss from selling a car
def calc_car_profit(purchase_price, years, annual_depreciation_rate, selling_price):
    """
    purchase_price: price you paid for the car
    years: number of years you owned the car
    annual_depreciation_rate: yearly depreciation rate (e.g., 0.15 = 15%)
    selling_price: actual selling price
    """

    # Residual value (theoretical current value of the car)
    residual_value = purchase_price * ((1 - annual_depreciation_rate) ** years)

    # Profit or loss (selling price - residual value)
    profit = selling_price - residual_value

    return residual_value, profit


# ---- Sample Simulation ----
purchase_price = 3_000_000      # Purchased for 3,000,000 yen
years = 5                       # Owned for 5 years
annual_dep = 0.15               # 15% depreciation per year
selling_price = 1_200_000       # Sold for 1,200,000 yen

residual_value, profit = calc_car_profit(
    purchase_price, years, annual_dep, selling_price
)

print(f"Residual value: {residual_value:,.0f} yen")
print(f"Profit (or loss): {profit:,.0f} yen")
```

```python
# 車の売却益・損を計算する関数
def calc_car_profit(purchase_price, years, annual_depreciation_rate, selling_price):
    """
    purchase_price: 購入価格
    years: 使用年数
    annual_depreciation_rate: 年間の価値減少率 (例: 0.15 = 15%)
    selling_price: 実際の売却価格
    """

    # 残存価値（現在の車の理論価格）
    residual_value = purchase_price * ((1 - annual_depreciation_rate) ** years)

    # 売却益（売却価格 - 残存価値）
    profit = selling_price - residual_value

    return residual_value, profit


# ---- サンプルシミュレーション ----
purchase_price = 3_000_000      # 300万円で購入
years = 5                       # 5年乗った
annual_dep = 0.15               # 毎年15%価値が下がる
selling_price = 1_200_000       # 120万円で売れた

residual_value, profit = calc_car_profit(
    purchase_price, years, annual_dep, selling_price
)

print(f"理論上の残存価値: {residual_value:,.0f} 円")
print(f"売却益（損）: {profit:,.0f} 円")
```
