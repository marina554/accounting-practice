```python
# ----------------------------------------
# (Visualizing the changes in the investment account and cash balance)
# ----------------------------------------

import pandas as pd
import matplotlib.pyplot as plt

# --- Initial settings ---
# Set information about the investing company
# Example: A parent company invests 10 million yen in a subsidiary
# and holds a 40% ownership interest
investment_amount = 10_000_000  # Investment amount (JPY)
ownership_ratio = 0.4           # Ownership ratio (40%)

# Net income of the subsidiary (each year)
subsidiary_net_income = [2_000_000, 3_000_000, 2_500_000, 4_000_000]
years = ["2021", "2022", "2023", "2024"]

# --- Lists to store data ---
investment_balance = []  # Investment account balance over time
cash_balance = []        # Cash balance over time

# Initial year settings
investment = investment_amount  # Beginning balance of investment account
cash = 50_000_000               # Cash balance (assumed amount)

# --- Simulate yearly changes ---
for net_income in subsidiary_net_income:
    # Recognize income under the equity method
    equity_income = net_income * ownership_ratio

    # Add equity income to the investment account
    investment += equity_income

    # Add dividend received to cash
    # (Assume 30% of net income is distributed as dividends)
    dividend = net_income * 0.3
    received_dividend = dividend * ownership_ratio
    cash += received_dividend

    # Record values in lists
    investment_balance.append(investment)
    cash_balance.append(cash)

# --- Create a DataFrame ---
df = pd.DataFrame({
    "Year": years,
    "Investment Account": investment_balance,
    "Cash Balance": cash_balance
})

# --- Display as a table ---
print("üìä Changes in Investment Account and Cash Balance by Year")
print(df)

# --- Plot the data ---
plt.figure(figsize=(8, 5))
plt.plot(df["Year"], df["Investment Account"], marker='o', label="Investment Account")
plt.plot(df["Year"], df["Cash Balance"], marker='s', label="Cash Balance")

plt.title("Changes in Investment Account and Cash Balance under the Equity Method", fontsize=13)
plt.xlabel("Year")
plt.ylabel("Amount (JPY)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
