```python

# -----------------------------------------
# Simple calculation of deferred tax assets/liabilities
# using temporary differences.
# Example: Difference in depreciation (book vs. tax)
# -----------------------------------------

# Book depreciation (e.g., straight-line method)
accounting_depr = 800_000

# Tax depreciation (e.g., declining-balance method)
tax_depr = 1_200_000

# Effective tax rate
tax_rate = 0.30

# Temporary difference
# If positive → future taxable income decreases → Deferred Tax Asset (DTA)
# If negative → future taxable income increases → Deferred Tax Liability (DTL)
temporary_difference = accounting_depr - tax_depr

# Amount of deferred tax
deferred_tax = temporary_difference * tax_rate

print("Book depreciation:", accounting_depr)
print("Tax depreciation:", tax_depr)
print("Temporary difference:", temporary_difference)
print("Tax rate:", tax_rate)

# Classification
if temporary_difference > 0:
    print("→ Deferred Tax Asset (DTA):", deferred_tax)
elif temporary_difference < 0:
    print("→ Deferred Tax Liability (DTL):", deferred_tax)
else:
    print("→ No temporary difference.")
```

```python
# -----------------------------------------
# 一時差異を使った繰延税金資産・負債のシンプル計算
# 題材：会計上と税務上の減価償却費の違い
# -----------------------------------------

# 会計上の減価償却費（例：定額法）
accounting_depr = 800_000

# 税務上の減価償却費（例：定率法）
tax_depr = 1_200_000

# 税率（法人税等実効税率）
tax_rate = 0.30

# 一時差異（temporary difference）
# マイナスなら「将来、会計利益の方が大きくなる」＝繰延税金負債
# プラスなら「将来、税務上費用が大きくなる」＝繰延税金資産
temporary_difference = accounting_depr - tax_depr

# 繰延税金の金額（DTA/DTL）
deferred_tax = temporary_difference * tax_rate

print("会計上の減価償却費:", accounting_depr)
print("税務上の減価償却費:", tax_depr)
print("一時差異:", temporary_difference)
print("税率:", tax_rate)

# 判定して表示
if temporary_difference > 0:
    print("→ 繰延税金資産（DTA）の発生:", deferred_tax)
elif temporary_difference < 0:
    print("→ 繰延税金負債（DTL）の発生:", deferred_tax)
else:
    print("→ 一時差異はありません。")
```
