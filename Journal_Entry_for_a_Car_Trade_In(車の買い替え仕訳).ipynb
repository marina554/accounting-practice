```python
def car_trade_in_entry(
    purchase_price_old,     # Acquisition cost of the old car
    accum_dep,              # Accumulated depreciation
    selling_price,          # Sale price of the old car
    purchase_price_new,     # Purchase price of the new car
    payment_method="cash"   # Payment method: "cash", "bank", etc.
):
    # Book value of the old car
    book_value = purchase_price_old - accum_dep

    # Profit or loss on disposal
    profit_or_loss = selling_price - book_value

    entries = []

    # --- Journal Entries for Disposing the Old Car ---
    # Receipt of cash from sale
    entries.append(("Cash", selling_price, "Debit"))

    # Reverse accumulated depreciation
    entries.append(("Accumulated Depreciation", accum_dep, "Debit"))

    # Remove the old vehicle from books
    entries.append(("Vehicle", purchase_price_old, "Credit"))

    # Gain or loss on disposal
    if profit_or_loss > 0:
        entries.append(("Gain on Sale of Fixed Assets", profit_or_loss, "Credit"))
    elif profit_or_loss < 0:
        entries.append(("Loss on Sale of Fixed Assets", -profit_or_loss, "Debit"))

    # --- Journal Entry for Purchasing the New Car ---
    entries.append(("Vehicle", purchase_price_new, "Debit"))

    if payment_method == "cash":
        entries.append(("Cash", purchase_price_new, "Credit"))
    elif payment_method == "bank":
        entries.append(("Bank", purchase_price_new, "Credit"))
    else:
        entries.append((payment_method, purchase_price_new, "Credit"))

    return entries



# ---- Example Simulation ----
old_price = 3_000_000      # Old car purchase cost
acc_dep = 2_000_000        # Accumulated depreciation
sell_price = 600_000       # Sale price
new_price = 4_000_000      # New car purchase cost

entries = car_trade_in_entry(old_price, acc_dep, sell_price, new_price)

for e in entries:
    print(f"{e[0]:30s}  {e[1]:>10,}  {e[2]}")
```

```python
def car_trade_in_entry(
    purchase_price_old,     # 旧車の取得原価
    accum_dep,              # 減価償却累計額
    selling_price,          # 売却額
    purchase_price_new,     # 新車の購入金額
    payment_method="cash"   # 支払方法: "cash", "bank", etc.
):
    # 帳簿価額
    book_value = purchase_price_old - accum_dep

    # 売却益 or 売却損
    profit_or_loss = selling_price - book_value

    entries = []

    # --- 旧車売却仕訳 ---
    # 現金の受取
    entries.append(("Cash", selling_price, "Debit"))

    # 減価償却累計
    entries.append(("Accumulated Depreciation", accum_dep, "Debit"))

    # 車両の除却
    entries.append(("Vehicle", purchase_price_old, "Credit"))

    # 売却益／損
    if profit_or_loss > 0:
        entries.append(("Gain on Sale of Fixed Assets", profit_or_loss, "Credit"))
    elif profit_or_loss < 0:
        entries.append(("Loss on Sale of Fixed Assets", -profit_or_loss, "Debit"))

    # --- 新車購入仕訳 ---
    entries.append(("Vehicle", purchase_price_new, "Debit"))

    if payment_method == "cash":
        entries.append(("Cash", purchase_price_new, "Credit"))
    elif payment_method == "bank":
        entries.append(("Bank", purchase_price_new, "Credit"))
    else:
        entries.append((payment_method, purchase_price_new, "Credit"))

    return entries



# ---- シミュレーション例 ----
old_price = 3_000_000      # 旧車取得原価
acc_dep = 2_000_000        # 減価償却累計額
sell_price = 600_000       # 売却金額
new_price = 4_000_000      # 新車購入価格

entries = car_trade_in_entry(old_price, acc_dep, sell_price, new_price)

for e in entries:
    print(f"{e[0]:30s}  {e[1]:>10,}  {e[2]}")
```
