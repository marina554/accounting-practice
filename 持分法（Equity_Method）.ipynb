```python
# =========================================
# 持分法の基本モデル（親会社1社＋関連会社1社）
# =========================================
import pandas as pd

# -----------------------------------------
# 1. 初期データ設定
# -----------------------------------------
# 親会社が関連会社株式を40%保有していると仮定します
ownership = 0.4  # 持分比率 40%

# 親会社の投資勘定（関連会社株式）
parent_investment = 1000  # 投資額（取得時点）

# 関連会社の当期純利益（例：1年間）
associate_net_income = 500  # 純利益

# 関連会社が親会社に配当を支払った金額
dividend_paid = 100  # 配当金（現金支払）

# -----------------------------------------
# 2. 親会社の持分に応じた利益の認識
# -----------------------------------------
# 関連会社の純利益 × 持分比率 = 親会社の持分利益
equity_income = associate_net_income * ownership

# 親会社はこの利益分だけ投資勘定を増やす（投資の価値が上がるため）
parent_investment += equity_income

print(f"親会社の投資勘定（持分利益反映後）: {parent_investment}")

# -----------------------------------------
# 3. 配当金を受け取った場合の処理
# -----------------------------------------
# 配当金は「投資先から現金を受け取る」ので、
# 投資勘定をその分だけ減額します（投資の回収）
parent_investment -= dividend_paid * ownership

print(f"配当金処理後の投資勘定: {parent_investment}")

# -----------------------------------------
# 4. 連結時の親会社BS上の表示イメージ
# -----------------------------------------
# （実務では、子会社の資産負債を取り込むのではなく、投資勘定を調整）
parent_bs = {
    "現金": 500 + dividend_paid * ownership,  # 配当金受取分、現金が増える
    "関連会社株式（投資勘定）": parent_investment,
    "純資産合計": 0  # 後で計算
}

# 純資産 = 資産合計
parent_bs["純資産合計"] = parent_bs["現金"] + parent_bs["関連会社株式（投資勘定）"]

# 表示用にDataFrameに変換
df_bs = pd.DataFrame(parent_bs, index=["親会社（持分法適用後）"])
print("\n親会社の貸借対照表（持分法適用後）:")
print(df_bs)

# -----------------------------------------
# 5. サマリー（説明出力）
# -----------------------------------------
print("\n【処理の流れ】")
print("① 関連会社の純利益500のうち40%（200）を親会社の投資勘定に加算。")
print("② 関連会社からの配当金100のうち40%（40）を受け取り、投資勘定を40減額。")
print("③ 結果として投資勘定は +160増加（1000→1160）。")
```
