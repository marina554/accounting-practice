```python
# ARO (Asset Retirement Obligation) Calculation

def calculate_aro(future_cost, discount_rate, years):
    """
    Discount future cost to present value.
    future_cost : Estimated future retirement cost
    discount_rate : Annual discount rate (e.g., 0.05 = 5%)
    years : Number of years until retirement
    """
    present_value = future_cost / ((1 + discount_rate) ** years)
    return present_value

def accretion_expense(pv_aro, discount_rate):
    """
    Annual Accretion Expense
    """
    return pv_aro * discount_rate


# ==== Example Usage ====
future_cost = 5_000_000     # Estimated future removal cost (yen)
discount_rate = 0.04        # Discount rate 4%
years = 10                  # Number of years until removal

pv_aro = calculate_aro(future_cost, discount_rate, years)
annual_accretion = accretion_expense(pv_aro, discount_rate)

print(f"Present Value of ARO (PV): {pv_aro:,.0f} yen")
print(f"Annual Accretion Expense:  {annual_accretion:,.0f} yen")
```

```python
# ARO (Asset Retirement Obligation) の計算

def calculate_aro(future_cost, discount_rate, years):
    """
    将来費用(future_cost)を現在価値に割引
    discount_rate : 年利率 (例: 0.05 = 5%)
    years : 年数
    """
    present_value = future_cost / ((1 + discount_rate) ** years)
    return present_value

def accretion_expense(pv_aro, discount_rate):
    """
    年次利息費用 (Accretion Expense)
    """
    return pv_aro * discount_rate


# ==== 使用例 ====
future_cost = 5000000      # 将来撤去費用（円）
discount_rate = 0.04       # 割引率 4%
years = 10                 # 使用後撤去までの期間 10年

pv_aro = calculate_aro(future_cost, discount_rate, years)
annual_accretion = accretion_expense(pv_aro, discount_rate)

print(f"資産除去債務の現在価値 (PV): {pv_aro:,.0f} 円")
print(f"年次利息費用 (Accretion Expense): {annual_accretion:,.0f} 円")
```
