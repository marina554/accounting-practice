```python
# ----------------------------------------
# æŒåˆ†æ³•ãƒãƒ¼ãƒˆãƒ•ã‚©ãƒªã‚ªï¼ˆæŠ•è³‡å‹˜å®šã¨ç¾é‡‘ã®æ¨ç§»ã‚’ã‚°ãƒ©ãƒ•åŒ–ï¼‰
# ----------------------------------------

import pandas as pd
import matplotlib.pyplot as plt

# --- åˆæœŸè¨­å®š ---
# æŠ•è³‡ä¼æ¥­ã®æƒ…å ±ã‚’è¨­å®šã—ã¾ã™
# ä¾‹ï¼šè¦ªä¼šç¤¾ãŒå­ä¼šç¤¾ã«1,000ä¸‡å††ã‚’å‡ºè³‡ã—ã€40%ã‚’ä¿æœ‰ã—ã¦ã„ã‚‹ã‚±ãƒ¼ã‚¹
investment_amount = 10_000_000  # æŠ•è³‡é¡ï¼ˆå††ï¼‰
ownership_ratio = 0.4           # æŒåˆ†æ¯”ç‡ï¼ˆ40%ï¼‰

# å­ä¼šç¤¾ã®ç´”åˆ©ç›Šï¼ˆå„å¹´åº¦ï¼‰
subsidiary_net_income = [2_000_000, 3_000_000, 2_500_000, 4_000_000]  # å„å¹´åº¦ã®ç´”åˆ©ç›Š
years = ["2021å¹´", "2022å¹´", "2023å¹´", "2024å¹´"]

# --- ãƒ‡ãƒ¼ã‚¿ã‚’æ ¼ç´ã™ã‚‹ãƒªã‚¹ãƒˆ ---
investment_balance = []  # æŠ•è³‡å‹˜å®šã®æ®‹é«˜æ¨ç§»
cash_balance = []        # ç¾é‡‘æ®‹é«˜æ¨ç§»

# åˆå¹´åº¦ã®è¨­å®š
investment = investment_amount  # æŠ•è³‡å‹˜å®šã®æœŸé¦–æ®‹é«˜
cash = 50_000_000               # ç¾é‡‘æ®‹é«˜ï¼ˆä»®ã®é‡‘é¡ï¼‰

# --- å¹´ã”ã¨ã®å¤‰åŒ–ã‚’ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³ ---
for net_income in subsidiary_net_income:
    # æŒåˆ†æ³•ã«ã‚ˆã‚‹åˆ©ç›Šã®èªè­˜
    equity_income = net_income * ownership_ratio

    # æŠ•è³‡å‹˜å®šã«æŒåˆ†åˆ©ç›Šã‚’åŠ ç®—
    investment += equity_income

    # ç¾é‡‘æ®‹é«˜ã«é…å½“æ”¯æ‰•ã„ï¼ˆã“ã“ã§ã¯ä»®ã«ç´”åˆ©ç›Šã®30%ã‚’é…å½“ã¨ã™ã‚‹ï¼‰
    dividend = net_income * 0.3
    received_dividend = dividend * ownership_ratio
    cash += received_dividend  # å—ã‘å–ã£ãŸé…å½“ã‚’ç¾é‡‘ã«åŠ ç®—

    # ãƒªã‚¹ãƒˆã«è¨˜éŒ²
    investment_balance.append(investment)
    cash_balance.append(cash)

# --- DataFrameã«ã¾ã¨ã‚ã‚‹ ---
df = pd.DataFrame({
    "å¹´åº¦": years,
    "æŠ•è³‡å‹˜å®š": investment_balance,
    "ç¾é‡‘æ®‹é«˜": cash_balance
})

# --- è¡¨å½¢å¼ã§å‡ºåŠ› ---
print("ğŸ“Š å¹´åº¦åˆ¥ã®æŠ•è³‡å‹˜å®šã¨ç¾é‡‘ã®æ¨ç§»")
print(df)

# --- ã‚°ãƒ©ãƒ•åŒ– ---
plt.figure(figsize=(8, 5))
plt.plot(df["å¹´åº¦"], df["æŠ•è³‡å‹˜å®š"], marker='o', label="æŠ•è³‡å‹˜å®š")
plt.plot(df["å¹´åº¦"], df["ç¾é‡‘æ®‹é«˜"], marker='s', label="ç¾é‡‘æ®‹é«˜")

plt.title("æŒåˆ†æ³•ã«ã‚ˆã‚‹æŠ•è³‡å‹˜å®šã¨ç¾é‡‘æ®‹é«˜ã®æ¨ç§»", fontsize=13)
plt.xlabel("å¹´åº¦")
plt.ylabel("é‡‘é¡ï¼ˆå††ï¼‰")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
